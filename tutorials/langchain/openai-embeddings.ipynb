```python
from langchain.document_loaders import TextLoader
from langchain_openai import OpenAIEmbeddings
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OpenAI embeddings."
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embeddings_1024
```

```python
query_result = embeddings_1024.embed_query(text)
len(query_result)
```

```python
loader = TextLoader("data/speech.txt")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splitted_docs = text_splitter.split_documents(docs)
splitted_docs
```

```python
## Vector Embedding and Vector StoreDB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(documents=splitted_docs, embedding=embeddings_1024)
db
```

```python
query = "It will be all the easier"
result = db.similarity_search(query)
print(result)
```
